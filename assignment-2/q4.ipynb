```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.exp(x) / ( (1 + x) ** (1 + x) )

# Define the range for x
x = np.linspace(-2, 2, 400)

# Compute the function values
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$\frac{e^x}{(1+x)^{1+x}}$', color='blue')
plt.title(r'Plot of $f(x) = \frac{e^x}{(1+x)^{1+x}}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
