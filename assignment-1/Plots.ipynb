<h3>Coxcombe Plots</h3>
<ul>
    <li>Interesting history: They were first used by Florence Nightingale to assess the number of soldiers who died every month and whether their death was caused by the wounds in the battlefield or diseases due to bad hygiene. The book which contained the charts was called coxcomb and the plots just inherited this name.</li>
    <li>They are modified pie charts in which each slice represents a parameter or an interval. The main difference between the two is that pie charts encode values using the angle subtended by the subsection but in coxcomb plots, each subsection subtends the same angle at the centre but the values are encoded by the fraction of area of the subsection covered by the plotted values</li>
</ul>

```python
import plotly.graph_objects as plotly

#representing the revenue from each venture of a company in the last eight months
fig = plotly.Figure()

fig.add_trace(plotly.Barpolar(
    r=[250,255,249,197,190,180,195,205],
    name='Retail (Non Luxury)'
))
fig.add_trace(plotly.Barpolar(
    r=[45,50,53,49,60,70,85,100],
    name='E-Commerce'
))
fig.add_trace(plotly.Barpolar(
    r=[80,84,89,94,97,113,120,97],
    name='Luxury Apparel'
))
fig.add_trace(plotly.Barpolar(
    r=[170,180,193,200,210,205,215,218],
    name='Consultancy Services'
))

fig.update_traces(text=['Dec 23', 'Jan 24', 'Feb 24', 'Mar 24', 'Apr 24', 'May 24', 'Jun 24', 'Jul 24'])
fig.update_layout(
    title='Revenue Distribution',
    width=600,
    height=600,
    font_size=16,
    legend_font_size=16,
)
fig.show()
```

```python
from matplotlib.ticker import PercentFormatter

df = pd.DataFrame({'Energy Consumption': [128,176,200,155,105,136,147,160]})
df.index = ['City A', 'City B', 'City C', 'City D', 'City E', 'City F', 'City G', 'City H']
df = df.sort_values(by='Energy Consumption',ascending=False)
df["cumpercentage"] = df["Energy Consumption"].cumsum()/df["Energy Consumption"].sum()*100


fig, ax = plt.subplots()
ax.bar(df.index, df["Energy Consumption"], color="C0")
ax2 = ax.twinx()
ax2.plot(df.index, df["cumpercentage"], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
plt.title("Energy Consumption of Eight Cities in a month")
plt.show()
```

<h3>Waterfall Plot</h3>
<ul>
    <li>They are used to visualise the variation of 2 dimensional idea with time. They are often used to depict spectograms and in acoustic analysis.</li>
    <li>Their name comes from the cascade or mountain like structures formed because of multiple curves being staggered across and vertically on the screen which causes the curves nearer to the viewer to mask those behind.</li>
</ul>

```python
points = 80
signals = 80
time = np.linspace(0, 10, points)

df = pd.DataFrame()

for i in range(signals):
    freq = 0.5 + i * 0.05  
    amp = 0.05*(1 + i * 0.1)     
    signal = amp * np.sin(2 * np.pi * freq * time)
    df[f"Signal_{i}"] = signal

df['Time'] = time
df.set_index('Time', inplace=True)
df.head()

fig = plotly.Figure()

# Use a colormap to assign different colors
colors = cm.rainbow(np.linspace(0, 1, len(df.columns)))

for i, (column, color) in enumerate(zip(df.columns, colors)):
    # Convert RGB to hex
    hex_color = '#%02x%02x%02x' % (int(color[0]*255), int(color[1]*255), int(color[2]*255))
    fig.add_trace(plotly.Scatter3d(
        x=df.index,
        y=np.full_like(df.index, i),
        z=df[column],
        mode='lines',
        line=dict(color=hex_color),
        name=f'Signal {i}'
    ))

fig.update_layout(
    scene=dict(
        xaxis_title='Time',
        yaxis_title='Signal Number',
        zaxis_title='amp'
    ),
    height=700,
    width=900,
    title='Waterfall Plot Example (dummy signals)'
)

```
