```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
m = 5
n = 20

# Create arrays of j values for the sum expressions
j_values = np.arange(m+1, n+1)

# Calculate the heights for the histograms
heights = 1 / (j_values - 1)

# Define the function 1/(x-1)
def func(x):
    return 1 / (x-1)

# Generate x values for plotting the function 1/(x-1)
x_values = np.linspace(m, n, 500)
y_values = func(x_values)

# First Histogram: From j-2 to j-1 with height 1/(j-1)
plt.figure(figsize=(10, 6))
for j, h in zip(j_values, heights):
    plt.bar(j-1, h, width=1, alpha=0.5, align='edge', color='blue', edgecolor='black', label=f'Interval [{j-2}, {j-1}]' if j == m+1 else "")
plt.plot(x_values, y_values, color='red', label=r'$f(x) = \frac{1}{x-1}$')
plt.title('Upper Bounded Graph')
plt.xlabel('x')
plt.ylabel('Height \( \frac{1}{j-1} \)')
plt.legend()
plt.grid(True)
plt.show()

# Second Histogram: From j-1 to j with height 1/(j-1)
# Modify the range to start one unit later and end one unit later
x_values_later = np.linspace(m+1, n+1, 500)
y_values_later = func(x_values_later)

plt.figure(figsize=(10, 6))
for j, h in zip(j_values, heights):
    plt.bar(j, h, width=1, alpha=0.5, align='edge', color='green', edgecolor='black', label=f'Interval [{j-1}, {j}]' if j == m+1 else "")
plt.plot(x_values_later, y_values_later, color='orange', label=r'$f(x) = \frac{1}{x-1}$')
plt.title('Lower Bounded Graph (Shifted by One Unit)')
plt.xlabel('x')
plt.ylabel('Height \( \frac{1}{j-1} \)')
plt.legend()
plt.grid(True)
plt.show()
```
