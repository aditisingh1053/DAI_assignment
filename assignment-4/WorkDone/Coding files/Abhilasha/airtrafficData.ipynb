```python
y=df['PASSENGERS CARRIED']
mae_scores = []
train_window = 60  # Adjust this window size as needed for rolling training set

for i in range(train_window, len(y)):
    # Use a rolling window to fit the model
    train_data = y[:i]
    model = ARIMA(train_data, order=(1, 1, 1))
    fit_model = model.fit()
    
    # Forecast the next step and calculate MAE against actual value
    forecast = fit_model.forecast(steps=1)
    mae_scores.append(mean_absolute_error([y[i]], forecast))

# Calculate the average MAE over the cross-validation
avg_mae = np.mean(mae_scores)
# print("Cross-Validation MAE:", avg_mae)
```
